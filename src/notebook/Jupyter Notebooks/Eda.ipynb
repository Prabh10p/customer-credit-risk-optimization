```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 14))

# 1. Age Distribution
plt.subplot(3, 3, 1)
sns.histplot(data=df, x="person_age", hue="loan_status", bins=10, kde=True, palette="Set1")
plt.title("Age Distribution by Loan Status")
plt.xlabel("Age")
plt.ylabel("Frequency")

# 2. Person Income Distribution
plt.subplot(3, 3, 2)
sns.histplot(data=df, x="person_income", hue="loan_status", bins=10,kde=True, color="purple")
plt.title("Person Income Distribution")
plt.xlabel("Person Income")
plt.ylabel("Frequency")

# 3. Employment Duration Distribution
plt.subplot(3, 3, 3)
sns.histplot(data=df, x="employement_duration", hue="loan_status", bins=10,kde = True, color="blue")
plt.title("Employment Duration Distribution")
plt.xlabel("Employment Duration")
plt.ylabel("Frequency")

# 4. Loan Amount Distribution
plt.subplot(3, 3, 4)
sns.histplot(data=df, x="loan_amount", hue="loan_status", bins=10,kde=True, color="orange")
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")

# 5. Loan Interest Rate Distribution
plt.subplot(3, 3, 5)
sns.histplot(data=df, x="loan_int_rate", hue="loan_status", bins=10,kde =True, color="orange")
plt.title("Loan Interest Rate Distribution")
plt.xlabel("Loan Interest Rate")
plt.ylabel("Frequency")

# 6. Loan Percent Income
plt.subplot(3, 3, 6)
sns.histplot(data=df, x="loan_percent_income", hue="loan_status", bins=10,kde =True, color="orange")
plt.title("Loan Percent Income")
plt.xlabel("Loan Percent")
plt.ylabel("Frequency")

# 7. Credit History Duration
plt.subplot(3, 3, 7)
sns.histplot(data=df, x="credit_history_duration", hue="loan_status", bins=10,kde=True, color="orange")
plt.title("Credit History Duration")
plt.xlabel("Credit History Length")
plt.ylabel("Frequency")

# Ensure layout is clean
plt.tight_layout()
plt.show()
```
